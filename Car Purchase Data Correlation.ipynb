#### Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
from matplotlib.pyplot import figure

plt.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

#### Load Dataset 

```python
df = pd.read_excel(r"C:\Users\asifs\Desktop\Python & Pandas Files\car_purchasing.xlsx")
df.head(10)
```

#### Data Exploration & Cleaning

```python
# checking for null values
df.isnull().sum()
```

```python
# checking data types
df.dtypes
```

```python
# fixing number format
df['age'] = df['age'].astype('int64')
df['annual Salary'] = df['annual Salary'].astype('int64')
df['credit card debt'] = df['credit card debt'].astype('int64')
df['net worth'] = df['net worth'].astype('int64')
df['car purchase amount'] = df['car purchase amount'].astype('int64')
```

```python
# fixing gender column
df['gender'] = df['gender'].replace(0, 'F')
df['gender'] = df['gender'].replace(1, 'M')
```

```python
# dropping duplicates
df.drop_duplicates()
```

#### Data Visualization using Scatter Plot

```python
sns.regplot(x = df['age'], y = df['car purchase amount'], scatter_kws = {'color' : 'red'}, line_kws = {'color' : 'blue'})
plt.title("Customer Age vs Car Purchase Amount")
plt.xlabel("Age")
plt.ylabel("Amount")
plt.show()
```

#### Viewing Correlations within the Dataset

```python
# viewing the correlation matrix
df.corr()
```

```python
# viewing the correlation matrix in heatmap form
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Numeric Features")
plt.show()
```

```python
# converting all the data into numerical format
df_numerical = df
for col_name in df_numerical.columns:
    if df_numerical[col_name].dtype == 'object':
        df_numerical[col_name] = df_numerical[col_name].astype('category')
        df_numerical[col_name] = df_numerical[col_name].cat.codes
        
df_numerical
```

```python
# viewing the correlation matrix of the numerical dataset in heatmap form
correlation_matrix = df_numerical.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# correlation pairs
corr_pairs = correlation_matrix.unstack()
pd.set_option('display.max.rows', 90)
corr_pairs
```

```python
# correlation pairs sorted
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# viewing features with the highest correlation
high_correlation = sorted_pairs[(sorted_pairs) > 0.4]
high_correlation
```
